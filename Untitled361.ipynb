```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import backend as K

import pyswarms as ps

# =========================================================
# PATHS
# =========================================================
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Lake Thermal Pollution Detection"
SHP_PATH = os.path.join(
    BASE_DIR,
    "WFSServer",
    "DWA_WB_Ph1_West_Bengal.shp"
)

MODEL_DIR = os.path.join(BASE_DIR, "models")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")

os.makedirs(MODEL_DIR, exist_ok=True)
os.makedirs(RESULT_DIR, exist_ok=True)
os.makedirs(GRAPH_DIR, exist_ok=True)

# =========================================================
# LOAD SHAPEFILE
# =========================================================
gdf = gpd.read_file(SHP_PATH)

# =========================================================
# SIMULATED THERMAL FEATURES (Replace later with real data)
# =========================================================
np.random.seed(42)

gdf["surface_temp"] = np.random.uniform(22, 40, len(gdf))
gdf["water_depth"] = np.random.uniform(1, 20, len(gdf))
gdf["industry_distance"] = np.random.uniform(0.1, 12, len(gdf))

gdf["thermal_pollution_index"] = (
    0.6 * gdf["surface_temp"]
    - 0.25 * gdf["industry_distance"]
    + 0.15 * gdf["water_depth"]
    + np.random.normal(0, 1, len(gdf))
)

FEATURES = ["surface_temp", "water_depth", "industry_distance"]
TARGET = "thermal_pollution_index"

X = gdf[FEATURES].values
y = gdf[TARGET].values.reshape(-1, 1)

# =========================================================
# SCALING
# =========================================================
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

joblib.dump(scaler_X, os.path.join(MODEL_DIR, "scaler_X.pkl"))
joblib.dump(scaler_y, os.path.join(MODEL_DIR, "scaler_y.pkl"))

# CNN expects (samples, timesteps, channels)
X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=42
)

# =========================================================
# CNN MODEL (SAFE VERSION)
# =========================================================
def build_model(filters, kernel_size):
    K.clear_session()  # VERY IMPORTANT for PSO

    filters = int(filters)
    kernel_size = int(kernel_size)

    # Ensure kernel never exceeds input length
    kernel_size = min(kernel_size, X_train.shape[1])

    model = Sequential([
        Conv1D(
            filters=filters,
            kernel_size=kernel_size,
            activation="relu",
            padding="same",
            input_shape=(X_train.shape[1], 1)
        ),
        GlobalMaxPooling1D(),  # FIXES negative dimension error
        Dense(32, activation="relu"),
        Dense(1)
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss="mse"
    )
    return model

# =========================================================
# PSO FITNESS FUNCTION
# =========================================================
def fitness_function(particles):
    losses = []

    for particle in particles:
        model = build_model(particle[0], particle[1])

        model.fit(
            X_train, y_train,
            epochs=8,
            batch_size=16,
            verbose=0
        )

        loss = model.evaluate(X_test, y_test, verbose=0)
        losses.append(loss)

    return np.array(losses)

# =========================================================
# PSO OPTIMIZATION
# =========================================================
options = {"c1": 1.5, "c2": 1.5, "w": 0.7}

bounds = (
    np.array([8, 1]),                 # filters, kernel
    np.array([64, X_train.shape[1]])
)

optimizer = ps.single.GlobalBestPSO(
    n_particles=8,
    dimensions=2,
    options=options,
    bounds=bounds
)

best_cost, best_pos = optimizer.optimize(fitness_function, iters=5)

# =========================================================
# FINAL MODEL TRAINING
# =========================================================
final_model = build_model(best_pos[0], best_pos[1])

history = final_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    callbacks=[EarlyStopping(patience=6, restore_best_weights=True)],
    verbose=1
)

# =========================================================
# SAVE MODEL (.h5)
# =========================================================
final_model.save(os.path.join(MODEL_DIR, "lake_thermal_cnn_pso.h5"))

# =========================================================
# EVALUATION
# =========================================================
y_pred = final_model.predict(X_test)

y_test_inv = scaler_y.inverse_transform(y_test)
y_pred_inv = scaler_y.inverse_transform(y_pred)

rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
r2 = r2_score(y_test_inv, y_pred_inv)

# =========================================================
# SAVE RESULTS (.json)
# =========================================================
results = {
    "rmse": float(rmse),
    "r2_score": float(r2),
    "best_filters": int(best_pos[0]),
    "best_kernel_size": int(best_pos[1]),
    "model": "CNN + PSO",
    "region": "West Bengal"
}

with open(os.path.join(RESULT_DIR, "thermal_pollution_results.json"), "w") as f:
    json.dump(results, f, indent=4)

# =========================================================
# SAVE CONFIG (.yaml)
# =========================================================
config = {
    "features": FEATURES,
    "target": TARGET,
    "optimizer": "PSO",
    "epochs": 50,
    "cnn": {
        "filters": int(best_pos[0]),
        "kernel_size": int(best_pos[1])
    }
}

with open(os.path.join(BASE_DIR, "config.yaml"), "w") as f:
    yaml.dump(config, f)

# =========================================================
# SAVE PREDICTIONS (.csv)
# =========================================================
pred_df = pd.DataFrame({
    "actual": y_test_inv.flatten(),
    "predicted": y_pred_inv.flatten()
})

pred_df.to_csv(
    os.path.join(RESULT_DIR, "thermal_pollution_predictions.csv"),
    index=False
)

# =========================================================
# TRAINING LOSS GRAPH
# =========================================================
plt.figure(figsize=(7, 5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.title("CNN + PSO Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.tight_layout()
plt.savefig(os.path.join(GRAPH_DIR, "training_loss.png"))
plt.show()

print("✅ PIPELINE COMPLETED SUCCESSFULLY")
```
