```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import backend as K

import pyswarms as ps

# =========================================================
# PATHS
# =========================================================
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Lake Thermal Pollution Detection"
SHP_PATH = os.path.join(
    BASE_DIR, "WFSServer", "DWA_WB_Ph1_West_Bengal.shp"
)

MODEL_DIR = os.path.join(BASE_DIR, "models")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")

os.makedirs(MODEL_DIR, exist_ok=True)
os.makedirs(RESULT_DIR, exist_ok=True)
os.makedirs(GRAPH_DIR, exist_ok=True)

# =========================================================
# LOAD SHAPEFILE
# =========================================================
gdf = gpd.read_file(SHP_PATH)

# =========================================================
# SIMULATED THERMAL FEATURES (replace with real data later)
# =========================================================
np.random.seed(42)

gdf["surface_temp"] = np.random.uniform(22, 40, len(gdf))
gdf["water_depth"] = np.random.uniform(1, 20, len(gdf))
gdf["industry_distance"] = np.random.uniform(0.1, 12, len(gdf))

gdf["thermal_pollution_index"] = (
    0.6 * gdf["surface_temp"]
    - 0.25 * gdf["industry_distance"]
    + 0.15 * gdf["water_depth"]
    + np.random.normal(0, 1, len(gdf))
)

FEATURES = ["surface_temp", "water_depth", "industry_distance"]
TARGET = "thermal_pollution_index"

X = gdf[FEATURES].values
y = gdf[TARGET].values.reshape(-1, 1)

# =========================================================
# SCALING
# =========================================================
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

joblib.dump(scaler_X, os.path.join(MODEL_DIR, "scaler_X.pkl"))
joblib.dump(scaler_y, os.path.join(MODEL_DIR, "scaler_y.pkl"))

X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=42
)

# =========================================================
# CNN MODEL (SAFE FOR PSO)
# =========================================================
def build_model(filters, kernel):
    K.clear_session()

    filters = int(filters)
    kernel = int(kernel)
    kernel = min(kernel, X_train.shape[1])

    model = Sequential([
        Conv1D(
            filters=filters,
            kernel_size=kernel,
            activation="relu",
            padding="same",
            input_shape=(X_train.shape[1], 1)
        ),
        GlobalMaxPooling1D(),
        Dense(32, activation="relu"),
        Dense(1)
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(0.001),
        loss="mse"
    )
    return model

# =========================================================
# PSO FITNESS FUNCTION
# =========================================================
def fitness_function(particles):
    losses = []

    for p in particles:
        model = build_model(p[0], p[1])
        model.fit(
            X_train, y_train,
            epochs=8,
            batch_size=16,
            verbose=0
        )
        loss = model.evaluate(X_test, y_test, verbose=0)
        losses.append(loss)

    return np.array(losses)

# =========================================================
# PSO OPTIMIZATION
# =========================================================
options = {"c1": 1.5, "c2": 1.5, "w": 0.7}

bounds = (
    np.array([8, 1]),
    np.array([64, X_train.shape[1]])
)

optimizer = ps.single.GlobalBestPSO(
    n_particles=8,
    dimensions=2,
    options=options,
    bounds=bounds
)

best_cost, best_pos = optimizer.optimize(fitness_function, iters=5)

# =========================================================
# FINAL MODEL TRAINING
# =========================================================
final_model = build_model(best_pos[0], best_pos[1])

history = final_model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    callbacks=[EarlyStopping(patience=6, restore_best_weights=True)],
    verbose=1
)

final_model.save(os.path.join(MODEL_DIR, "lake_thermal_cnn_pso.h5"))

# =========================================================
# EVALUATION
# =========================================================
y_pred = final_model.predict(X_test)

y_test_inv = scaler_y.inverse_transform(y_test)
y_pred_inv = scaler_y.inverse_transform(y_pred)

rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
r2 = r2_score(y_test_inv, y_pred_inv)

# =========================================================
# SAVE RESULTS
# =========================================================
json.dump(
    {
        "rmse": float(rmse),
        "r2_score": float(r2),
        "best_filters": int(best_pos[0]),
        "best_kernel": int(best_pos[1])
    },
    open(os.path.join(RESULT_DIR, "results.json"), "w"),
    indent=4
)

yaml.dump(
    {
        "features": FEATURES,
        "target": TARGET,
        "optimizer": "PSO",
        "epochs": 50
    },
    open(os.path.join(BASE_DIR, "config.yaml"), "w")
)

pd.DataFrame({
    "actual": y_test_inv.flatten(),
    "predicted": y_pred_inv.flatten()
}).to_csv(
    os.path.join(RESULT_DIR, "predictions.csv"),
    index=False
)

# =========================================================
# ===================== GRAPHS ============================
# =========================================================

# 1️⃣ Accuracy Graph
plt.figure(figsize=(6,4))
sns.barplot(x=["RMSE", "R2"], y=[rmse, r2])
plt.title("Model Accuracy Metrics")
plt.savefig(os.path.join(GRAPH_DIR, "accuracy_graph.png"))
plt.show()

# 2️⃣ Comparison Graph
plt.figure(figsize=(6,6))
plt.scatter(y_test_inv, y_pred_inv, alpha=0.6)
plt.plot(
    [y_test_inv.min(), y_test_inv.max()],
    [y_test_inv.min(), y_test_inv.max()],
    "r--"
)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.savefig(os.path.join(GRAPH_DIR, "comparison_graph.png"))
plt.show()

# 3️⃣ Heatmap
plt.figure(figsize=(6,5))
sns.heatmap(
    gdf[FEATURES + [TARGET]].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Feature Correlation Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "heatmap.png"))
plt.show()

# 4️⃣ Result Error Graph
errors = y_test_inv.flatten() - y_pred_inv.flatten()
plt.figure(figsize=(6,4))
sns.histplot(errors, bins=20, kde=True)
plt.title("Prediction Error Distribution")
plt.savefig(os.path.join(GRAPH_DIR, "result_graph.png"))
plt.show()

# 5️⃣ Prediction Distribution Graph
plt.figure(figsize=(6,4))
sns.kdeplot(y_test_inv.flatten(), label="Actual", fill=True)
sns.kdeplot(y_pred_inv.flatten(), label="Predicted", fill=True)
plt.legend()
plt.title("Prediction Distribution")
plt.savefig(os.path.join(GRAPH_DIR, "prediction_graph.png"))
plt.show()

print("✅ ALL RESULTS & GRAPHS GENERATED SUCCESSFULLY")
```
